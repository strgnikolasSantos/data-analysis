```python

import pandas as pd
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from webdriver_manager.microsoft import EdgeChromiumDriverManager

# Configura o serviço do Edge WebDriver
service = Service(EdgeChromiumDriverManager().install())

# Inicializa o driver do Edge
driver = webdriver.Edge(service=service)
```

```python
#Entrando no site
url = 'https://www.fundamentus.com.br/resultado.php'
driver.get(url)
```

```python
local_tabela = '/html/body/div[1]/div[2]/table'
elemento = driver.find_element("xpath",local_tabela)

html_tabela = elemento.get_attribute("outerHTML")

html_tabela
```

```python
tabela = pd.read_html(str(html_tabela), thousands=".", decimal=",")[0]
tabela
```

```python
driver.quit()
```

```python
#Tratamento da taela
tabela = tabela.set_index("Papel")

tabela = tabela[["Cotação","EV/EBITDA","ROIC","Liq.2meses"]]

display(tabela)
```

```python
tabela["ROIC"] = tabela["ROIC"].str.replace("%","")
tabela["ROIC"] = tabela["ROIC"].str.replace(".","")
tabela["ROIC"] = tabela["ROIC"].str.replace(",",".")
tabela["ROIC"] = tabela["ROIC"].astype(float)

display(tabela)
```

```python
#Crianddo a Carteira
tabela = tabela[tabela["Liq.2meses"]> 1000000] 
display(tabela)
```

```python
tabela = tabela[tabela["EV/EBITDA"] > 0] 
tabela = tabela[tabela["Liq.2meses"] > 1000000] 
display(tabela)
```

```python
tabela["ranking_ev_ebit"] = tabela["EV/EBITDA"].rank(ascending=True)
tabela["ranking_roic"] = tabela["ROIC"].rank(ascending=False)

tabela["ranking_total"] = tabela["ranking_roic"] + tabela["ranking_ev_ebit"]

tabela = tabela.sort_values("ranking_total")
tabela.head(10)

display(tabela)

tabela = tabela.head(10)

tickers = tabela.index
```
